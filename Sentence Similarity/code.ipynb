```python
def cleaning(doc):
    txt = [token.lemma_ for token in doc if not token.is_stop]
    return ' '.join(txt)
```

```python
def feature_generator(feature1,feature2,score):
    sent_a_cs=[]
    for i in feature1:
        sent_a_cs.append(np.mean(i))
    sent_a_cs_mat = np.array(sent_a_cs)
    sent_b_cs=[]
    for i in feature2:
        sent_b_cs.append(np.mean(i))
    sent_b_cs_mat = np.array(sent_b_cs)
    X=[]
    ind=[]
    for i in range(len(sent_a_cs)):
        a = sent_a_cs[i]
        b = sent_b_cs[i]
        if not math.isnan(a) and not math.isnan(b):
            sub = [a,b]
            X.append(sub)
        else:
            ind.append(i)
    Y = list(score)
    for i in ind:
        del Y[i]
    return X,Y
    
```

```python

def get_feature_model1(data_frame):
    """
    Input a data frame and return the embedding vectors for the each sentence column using non_cont_model1,
    Return 2 matrices each of shape (#_samples, #size_of_word_emb).
    """
    sent_a = list(data_frame["sent_a"])
    sent_a_mat =[]
    sent_b = list(data_frame["sent_b"])
    sent_b_mat =[]
    for i in sent_a:
        tokens = nlp(i)
        i_ = cleaning(tokens)
        i_ = re.sub("[^A-Za-z']+", ' ', str(i_)).lower()
        tokens = nlp(i_)
        sent_vec =[]
        for j in tokens:
            if j.text.upper() in non_cont_model1:
                sent_vec.append(non_cont_model1[j.text.upper()])
            elif j.text.lower() in non_cont_model1:
                sent_vec.append(non_cont_model1[j.text.lower()])
            else:
                continue
        sent_a_mat.append(sent_vec)
    for i in sent_b:
        tokens = nlp(i)
        i_ = cleaning(tokens)
        i_ = re.sub("[^A-Za-z']+", ' ', str(i_)).lower()
        tokens = nlp(i_)
        sent_vec =[]
        for j in tokens:
            if j.text.upper() in non_cont_model1:
                sent_vec.append(non_cont_model1[j.text.upper()])
            elif j.text.lower() in non_cont_model1:
                sent_vec.append(non_cont_model1[j.text.lower()])
            else:
                continue
        sent_b_mat.append(sent_vec)
    return np.array(sent_a_mat),np.array(sent_b_mat)


non_cont_model1 = KeyedVectors.load_word2vec_format(INPUT_PATH+NON_CONEXTUAL_MODEL_TYPE, binary=True, limit=100000)

feature_1_train, feature_2_train = get_feature_model1(df_train)
feature_1_dev, feature_2_dev = get_feature_model1(df_dev)
feature_1_test, feature_2_test = get_feature_model1(df_test)

X_train, Y_train = feature_generator(feature_1_train,feature_2_train, df_train['score'])
X_dev, Y_dev = feature_generator(feature_1_dev,feature_2_dev, df_dev['score'])
X_test, Y_test = feature_generator(feature_1_test,feature_2_test, df_test['score'])
# Initiate a regression model and train it.
regressor = RandomForestRegressor(**parameters)
regressor.fit(X_train, Y_train)   
# Print spearmanr correlation on the predicted output of the dev and test sets.
Y_pred_dev= regressor.predict(X_dev)
Y_pred_test= regressor.predict(X_test)
c_dev,p_dev = stats.spearmanr(Y_pred_dev,Y_dev)
c_test,p_test = stats.spearmanr(Y_pred_test,Y_test)
print("Correlation coefficient Dev Set:"+str(round(c_dev,3)))
print("Correlation coefficient Test Set:"+str(round(c_test,3)))
```
